## ✅ Task 1: Replace Hardcoded Prompts (10%)
**Goal:** Convert a hardcoded f-string prompt into a reusable `PromptTemplate`.

```python
!pip install -q langchain-core langchain
```

```python
from langchain_core.prompts import PromptTemplate, ChatPromptTemplate
from langchain_core.prompts import SystemMessagePromptTemplate, HumanMessagePromptTemplate

print("✅ LangChain components imported successfully!")
```

```python
# ─────────────────────────────────────────────────────────────
# TASK 1: Replace Hardcoded Prompts with PromptTemplate
# ─────────────────────────────────────────────────────────────

# ❌ Old approach using f-strings (hardcoded – NOT recommended)
# def explain_topic(topic):
#     return f"Explain {topic} in simple terms for beginners"

# ✅ New approach: LangChain PromptTemplate (reusable, dynamic)
explain_template = PromptTemplate(
    input_variables=["topic"],          # List of dynamic variables
    template="Explain {topic} in simple terms for beginners"
)

# ── Test the template with different topics ──
topics = ["Machine Learning", "Neural Networks", "Python"]

print("=" * 55)
print("TASK 1 OUTPUT – Explain Topic Template")
print("=" * 55)

for topic in topics:
    # .format() fills in the variable and returns the final prompt string
    prompt = explain_template.format(topic=topic)
    print(f"→ {prompt}")

print("\n✅ Task 1 Complete – No f-strings, fully reusable PromptTemplate!")
```

## ✅ Task 2: Multi-Input Prompt System (15%)
**Goal:** Build a template that accepts `topic`, `audience`, and `tone` as inputs.

```python
# ─────────────────────────────────────────────────────────────
# TASK 2: Multi-Input Prompt Template
# ─────────────────────────────────────────────────────────────

# Define a PromptTemplate with 3 input variables
multi_input_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone"
)

# ── Test cases as defined in the assignment ──
test_cases = [
    {"topic": "AI",            "audience": "beginners",  "tone": "friendly"},
    {"topic": "Python",        "audience": "kids",        "tone": "fun"},
    {"topic": "Deep Learning", "audience": "engineers",  "tone": "technical"},
]

print("=" * 60)
print("TASK 2 OUTPUT – Multi-Input Prompt System")
print("=" * 60)

for idx, case in enumerate(test_cases, start=1):
    # Unpack dictionary into the template using **kwargs
    prompt = multi_input_template.format(**case)
    print(f"Test {idx}: {prompt}")

print("\n✅ Task 2 Complete – 3-variable PromptTemplate working correctly!")
```

## ✅ Task 3: Prompt Variations Engine (15%)
**Goal:** Create 3 different PromptTemplates for Teaching, Interview, and Storytelling styles.

```python
# ─────────────────────────────────────────────────────────────
# TASK 3: Prompt Variations Engine
# ─────────────────────────────────────────────────────────────

# Store all templates in a dictionary for easy access and extensibility
prompt_variations = {
    "teaching":     PromptTemplate(
                        input_variables=["topic"],
                        template="Explain {topic} clearly step by step"
                    ),
    "interview":    PromptTemplate(
                        input_variables=["topic"],
                        template="Ask 3 questions about {topic}"
                    ),
    "storytelling": PromptTemplate(
                        input_variables=["topic"],
                        template="Explain {topic} as a story"
                    ),
}

# ── Run all 3 variations for the same topic ──
topic = "Machine Learning"

print("=" * 60)
print(f"TASK 3 OUTPUT – Prompt Variations for: '{topic}'")
print("=" * 60)

for style, template in prompt_variations.items():
    # Same topic, 3 different prompt outputs depending on style
    prompt = template.format(topic=topic)
    print(f"[{style.upper():12}] → {prompt}")

print("\n✅ Task 3 Complete – 3 distinct variations from 1 topic input!")
```

## ✅ Task 4: ChatPromptTemplate System (15%)
**Goal:** Build a role-aware chat prompt system using `System` and `Human` message templates.

```python
# ─────────────────────────────────────────────────────────────
# TASK 4: ChatPromptTemplate System
# ─────────────────────────────────────────────────────────────

# Role descriptions define the AI's behavior (system-level instruction)
role_descriptions = {
    "teacher":     "You are an expert teacher. Explain concepts clearly with examples.",
    "interviewer": "You are a professional interviewer. Ask insightful technical questions.",
    "motivator":   "You are an enthusiastic motivator. Inspire people to learn new topics.",
}

def build_chat_prompt(role: str, topic: str) -> list:
    """
    Builds a ChatPromptTemplate for the given role and topic.

    Args:
        role  (str): One of 'teacher', 'interviewer', 'motivator'
        topic (str): The subject to discuss

    Returns:
        list: A list of formatted chat messages (System + Human)
    """
    # Get the system instruction for the selected role
    system_instruction = role_descriptions.get(role, "You are a helpful assistant.")

    # Build a ChatPromptTemplate with a System message and a Human message
    chat_template = ChatPromptTemplate.from_messages([
        ("system", system_instruction),          # Defines AI behavior
        ("human", "Tell me about {topic}"),      # User's dynamic input
    ])

    # Format the template by substituting the topic variable
    messages = chat_template.format_messages(topic=topic)
    return messages

# ── Test with the example from the assignment ──
test_role  = "teacher"
test_topic = "Neural Networks"

print("=" * 60)
print("TASK 4 OUTPUT – ChatPromptTemplate System")
print("=" * 60)

# Run for all 3 roles with the same topic
for role in role_descriptions:
    messages = build_chat_prompt(role=role, topic=test_topic)
    print(f"\n🔹 Role: {role.upper()}")
    for msg in messages:
        print(f"   [{msg.type.upper():6}]: {msg.content}")

print("\n✅ Task 4 Complete – Role-aware ChatPromptTemplate working!")
```

## ✅ Task 5: Input Validation Layer (10%)
**Goal:** Validate user inputs before they are passed into templates.

```python
# ─────────────────────────────────────────────────────────────
# TASK 5: Input Validation Layer
# ─────────────────────────────────────────────────────────────

# Define allowed values for each parameter
VALID_AUDIENCES = ["beginner", "intermediate", "expert"]
VALID_TONES     = ["formal", "casual", "fun"]

# Default fallback values used when input is invalid
DEFAULT_AUDIENCE = "beginner"
DEFAULT_TONE     = "casual"


def validate_inputs(audience: str, tone: str) -> tuple:
    """
    Validates audience and tone inputs against allowed values.
    Raises a ValueError for invalid inputs and prints a warning,
    then falls back to safe defaults.

    Args:
        audience (str): Target audience level
        tone     (str): Desired communication tone

    Returns:
        tuple: (validated_audience, validated_tone)
    """
    validated_audience = audience
    validated_tone     = tone

    # ── Validate audience ──
    if audience.lower() not in VALID_AUDIENCES:
        print(f"⚠️  Invalid audience '{audience}'. "
              f"Allowed: {VALID_AUDIENCES}. "
              f"Defaulting to '{DEFAULT_AUDIENCE}'.")
        validated_audience = DEFAULT_AUDIENCE

    # ── Validate tone ──
    if tone.lower() not in VALID_TONES:
        print(f"⚠️  Invalid tone '{tone}'. "
              f"Allowed: {VALID_TONES}. "
              f"Defaulting to '{DEFAULT_TONE}'.")
        validated_tone = DEFAULT_TONE

    return validated_audience, validated_tone


# ── Test Cases ──
test_inputs = [
    ("beginner",  "formal"),    # ✅ Both valid
    ("kids",      "fun"),        # ❌ Invalid audience
    ("expert",    "rude"),       # ❌ Invalid tone
    ("student",   "angry"),      # ❌ Both invalid
]

print("=" * 60)
print("TASK 5 OUTPUT – Input Validation Layer")
print("=" * 60)

for audience, tone in test_inputs:
    print(f"\n📥 Input → audience='{audience}', tone='{tone}'")
    valid_audience, valid_tone = validate_inputs(audience, tone)
    print(f"✅ Used  → audience='{valid_audience}', tone='{valid_tone}'")

print("\n✅ Task 5 Complete – Validation with graceful defaults!")
```

## ✅ Task 6: Prompt Generator App (15%)
**Goal:** Create a `generate_prompt()` function combining validation + templates + styles.

```python
# ─────────────────────────────────────────────────────────────
# TASK 6: Prompt Generator App
# ─────────────────────────────────────────────────────────────

# ── Style-specific PromptTemplates ──
# Each style has a unique template structure
style_templates = {
    "teaching":     PromptTemplate(
                        input_variables=["topic", "audience", "tone"],
                        template=(
                            "Explain {topic} for {audience} in a {tone} teaching style, "
                            "step by step with clear examples."
                        )
                    ),
    "interview":    PromptTemplate(
                        input_variables=["topic", "audience", "tone"],
                        template=(
                            "Generate 3 insightful interview questions about {topic} "
                            "for a {audience}-level candidate using a {tone} tone."
                        )
                    ),
    "storytelling": PromptTemplate(
                        input_variables=["topic", "audience", "tone"],
                        template=(
                            "Explain {topic} for {audience} as an engaging story "
                            "in a {tone} tone."
                        )
                    ),
}


def generate_prompt(topic: str, audience: str, tone: str, style: str) -> str:
    """
    Full pipeline: validate inputs → select template → generate prompt.

    Args:
        topic    (str): Subject to explain
        audience (str): Target audience level
        tone     (str): Communication tone
        style    (str): One of 'teaching', 'interview', 'storytelling'

    Returns:
        str: The fully generated prompt string
    """
    # Step 1: Validate inputs (uses Task 5 function)
    validated_audience, validated_tone = validate_inputs(audience, tone)

    # Step 2: Select the correct template based on style
    if style not in style_templates:
        print(f"⚠️  Unknown style '{style}'. Defaulting to 'teaching'.")
        style = "teaching"

    selected_template = style_templates[style]

    # Step 3: Format and return the generated prompt
    return selected_template.format(
        topic=topic,
        audience=validated_audience,
        tone=validated_tone
    )


# ── Test Cases ──
test_runs = [
    {"topic": "Neural Networks",  "audience": "beginner",     "tone": "fun",    "style": "storytelling"},
    {"topic": "Data Structures",  "audience": "intermediate", "tone": "casual", "style": "teaching"},
    {"topic": "System Design",    "audience": "expert",       "tone": "formal", "style": "interview"},
    {"topic": "Python Basics",    "audience": "kids",         "tone": "fun",    "style": "teaching"},  # Invalid audience
]

print("=" * 65)
print("TASK 6 OUTPUT – Prompt Generator App")
print("=" * 65)

for i, run in enumerate(test_runs, start=1):
    print(f"\n--- Run {i} ---")
    result = generate_prompt(**run)
    print(f"📤 Generated Prompt:\n   {result}")

print("\n✅ Task 6 Complete – Full pipeline: Validation → Template → Output!")
```

## ✅ Task 7: Template Reusability Test (10%)
**Goal:** Prove that ONE template can generate 5 structurally identical but contextually different prompts.

```python
# ─────────────────────────────────────────────────────────────
# TASK 7: Template Reusability Test
# ─────────────────────────────────────────────────────────────

# ONE reusable template – used for ALL 5 test inputs
reusable_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} to a {audience} in a {tone} and engaging way."
)

# 5 different inputs to demonstrate reusability
five_inputs = [
    {"topic": "Blockchain",         "audience": "beginner",     "tone": "casual"},
    {"topic": "Quantum Computing",  "audience": "expert",       "tone": "formal"},
    {"topic": "Artificial Intelligence", "audience": "intermediate", "tone": "fun"},
    {"topic": "Cloud Computing",    "audience": "beginner",     "tone": "friendly"},
    {"topic": "Cybersecurity",      "audience": "expert",       "tone": "technical"},
]

print("=" * 65)
print("TASK 7 OUTPUT – Template Reusability Test (1 Template, 5 Outputs)")
print("=" * 65)
print(f"\n🔁 Template Used:\n   '{reusable_template.template}'\n")

for idx, inputs in enumerate(five_inputs, start=1):
    # Same template, different data → different output each time
    prompt = reusable_template.format(**inputs)
    print(f"Output {idx}: {prompt}")

print("\n✅ Task 7 Complete – Same structure, 5 unique outputs demonstrated!")
```
